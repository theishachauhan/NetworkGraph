```python
import networkx as nx
import matplotlib.pyplot as plt

# use reference https://networkx.org/documentation/stable/reference/generators.html to change the generators
```

```python
# Function to plot any NetworkX graph on Spring layout
def plot_graph_spring(G, path=None, title="Graph Visualization"):
    print("\n\n")
    pos = nx.spring_layout(G, seed=712)  # Consistent layout for different graphs
    plt.figure(figsize=(6, 4))
    nx.draw(G, pos, with_labels=True, node_size=100, node_color="skyblue", edge_color="gray")

    # Highlight path if provided
    if path:
        path_edges = list(zip(path, path[1:]))
        nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color="red", width=2)

    plt.title(title)
    plt.show()
```

```python
# Function to plot any NetworkX graph on Kamada Kawai Layout
def plot_graph_kamada(G, path=None, title="Graph Visualization"):
    print("\n\n")
    pos = nx.kamada_kawai_layout(G)
    plt.figure(figsize=(6, 4))
    nx.draw(G, pos, with_labels=True, node_size=100, node_color="skyblue", edge_color="gray")

    # Highlight path if provided
    if path:
        path_edges = list(zip(path, path[1:]))
        nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color="red", width=2)

    plt.title(title)
    plt.show()
```

```python
# Function to plot any NetworkX graph on Spectral layout
def plot_graph_Spectral(G, path=None, title="Graph Visualization"):
    print("\n\n")
    pos = nx.spectral_layout(G)
    plt.figure(figsize=(6, 4))
    nx.draw(G, pos, with_labels=True, node_size=100, node_color="skyblue", edge_color="gray")

    # Highlight path if provided
    if path:
        path_edges = list(zip(path, path[1:]))
        nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color="red", width=2)

    plt.title(title)
    plt.show()
```

```python
# Create different types of graphs

#nx.fast_gnp_random_graph(50, 0.02)
# 50 is the number of nodes and 0.02 is the edge probability

random_graph = nx.fast_gnp_random_graph(50, 0.02)
lattice_graph = nx.grid_2d_graph(5, 10)
forest_fire_graph = nx.gnp_random_graph(50, 0.15)  # Placeholder for a "forest fire" graph
erdos_renyi_graph = nx.erdos_renyi_graph(50, 0.08)
```

```python
# Plot the graphs
plot_graph_Spectral(random_graph, title="Random Graph")
plot_graph_Spectral(lattice_graph, title="Lattice Graph")
plot_graph_Spectral(forest_fire_graph, title="Forest Fire Graph")
plot_graph_Spectral(erdos_renyi_graph, title="Erdos-Renyi Graph")
```

```python
plot_graph_kamada(random_graph, title="Random Graph")
plot_graph_kamada(lattice_graph, title="Lattice Graph")
plot_graph_kamada(forest_fire_graph, title="Forest Fire Graph")
plot_graph_kamada(erdos_renyi_graph, title="Erdos-Renyi Graph")
```

```python
plot_graph_spring(random_graph, title="Random Graph")
plot_graph_spring(lattice_graph, title="Lattice Graph")
plot_graph_spring(forest_fire_graph, title="Forest Fire Graph")
plot_graph_spring(erdos_renyi_graph, title="Erdos-Renyi Graph")
```
